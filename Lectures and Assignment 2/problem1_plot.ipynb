```python
import matplotlib.pyplot as plt

# Axis values for demand and their corresponding probabilities
demand = [0, 2000, 7000, 13000, 25000, 35000, 40000, 60000]
prob = [0, 0.1, 0.4, 0.5, 0.6, 0.7, 0.9, 1.0]  # Adjusted to maintain consistency

# Adjust the code to show ticks at the values of demand and center the grid on those points

# Create a plot with specified figure size
plt.figure(figsize=(11, 4))

# Plot the demand vs. probabilities with markers at data points
plt.plot(demand, prob, marker='o', linestyle='-')  # Use a solid line

# Set the title and labels of the plot
plt.title('CDF of Demand')
plt.xlabel('Demand')
plt.ylabel('Cumulative Probability')

# Set the grid to be displayed
plt.grid(True)

# Define the ticks on the x-axis to match the demand values
plt.xticks(demand)
plt.yticks(prob)

# Show the plot
plt.show()

```

```python
# Given demand levels and their corresponding CDF values
demand_levels = [0, 2000, 7000, 13000, 25000, 35000, 40000, 60000]
CDF_values = [0, 0.1, 0.4, 0.5, 0.6, 0.7, 0.9, 1.0]

# Find the expected demand
expected_demand = 0
for i in range(1, len(demand_levels)):
    expected_demand += demand_levels[i] * (CDF_values[i] - CDF_values[i - 1])

print(f"Expected demand = {expected_demand}")
```

```python


# Calculate the midpoints of each interval
midpoints = [(demand_levels[i] + demand_levels[i+1]) / 2 for i in range(len(demand_levels) - 1)]

# Calculate the probability for each interval
interval_probabilities = [CDF_values[i+1] - CDF_values[i] for i in range(len(CDF_values) - 1)]

# Compute the expected demand
expected_demand = sum(midpoint * probability for midpoint, probability in zip(midpoints, interval_probabilities))
expected_demand
```
