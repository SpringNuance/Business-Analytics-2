```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
profits_A = np.array([-10, 0, 10, 20, 20, 50])  # Profits from investment A
profits_B = np.array([-5, -5, 0, 30, 40, 70])  # Profits from investment B
probabilities = np.array([0.05, 0.15, 0.30, 0.35, 0.10, 0.05])  # Probabilities of each scenario

# Sort the profits for the cumulative distribution function (CDF)
sorted_indices_A = np.argsort(profits_A)
sorted_indices_B = np.argsort(profits_B)

sorted_profits_A = profits_A[sorted_indices_A]
sorted_probabilities_A = probabilities[sorted_indices_A]

sorted_profits_B = profits_B[sorted_indices_B]
sorted_probabilities_B = probabilities[sorted_indices_B]

# Calculate the CDF for both investments
cdf_A = np.cumsum(sorted_probabilities_A)
cdf_B = np.cumsum(sorted_probabilities_B)

# Find the indices where the profit is 20
indices_of_20 = np.where(sorted_profits_A == 20)[0]

# Sum the probabilities for profit 20
sum_prob_at_20 = np.sum(sorted_probabilities_A[indices_of_20])

# Remove the original entries for profit 20
sorted_profits_A = np.delete(sorted_profits_A, indices_of_20)
sorted_probabilities_A = np.delete(sorted_probabilities_A, indices_of_20)

# Add the combined entry for profit 20
sorted_profits_A = np.append(sorted_profits_A, 20)
sorted_probabilities_A = np.append(sorted_probabilities_A, sum_prob_at_20)

# Re-sort the arrays since we added a new entry
sorted_indices_A = np.argsort(sorted_profits_A)
sorted_profits_A = sorted_profits_A[sorted_indices_A]
sorted_probabilities_A = sorted_probabilities_A[sorted_indices_A]

# Recalculate the CDF for Investment A after adjustment
cdf_A = np.cumsum(sorted_probabilities_A)

# FInd the indices where the profit is -5
indices_of_neg5 = np.where(sorted_profits_B == -5)[0]

# Sum the probabilities for profit -5
sum_prob_at_neg5 = np.sum(sorted_probabilities_B[indices_of_neg5])

# Remove the original entries for profit -5
sorted_profits_B = np.delete(sorted_profits_B, indices_of_neg5)
sorted_probabilities_B = np.delete(sorted_probabilities_B, indices_of_neg5)

# Add the combined entry for profit -5
sorted_profits_B = np.append(sorted_profits_B, -5)
sorted_probabilities_B = np.append(sorted_probabilities_B, sum_prob_at_neg5)

# Re-sort the arrays since we added a new entry
sorted_indices_B = np.argsort(sorted_profits_B)
sorted_profits_B = sorted_profits_B[sorted_indices_B]
sorted_probabilities_B = sorted_probabilities_B[sorted_indices_B]

# Recalculate the CDF for Investment B after adjustment
cdf_B = np.cumsum(sorted_probabilities_B)


# Create a figure with two subplots
fig, ax = plt.subplots(2, 2, figsize=(14, 10))

# Plot the PDF for investment A
ax[0, 0].stem(sorted_profits_A, sorted_probabilities_A, 'b', markerfmt='bo', basefmt=" ", linefmt='b-', label='Investment A')
ax[0, 0].set_title('Probability Density Function (PDF)')
ax[0, 0].set_xlabel('Profits ($M)')
ax[0, 0].set_ylabel('Probability')
ax[0, 0].set_ylim(-0.01, 0.5)
ax[0, 0].legend(loc=2)

# Plot the CDF for investment A
ax[0, 1].step(sorted_profits_A, cdf_A, 'b', where='post', label='Investment A')
ax[0, 1].set_title('Cumulative Distribution Function (CDF)')
ax[0, 1].set_xlabel('Profits ($M)')
ax[0, 1].set_ylabel('Cumulative Probability')
ax[0, 1].set_ylim(-0.01, 1.1)
ax[0, 1].legend(loc=2)

# Plot the PDF for investment B
ax[1, 0].stem(sorted_profits_B, sorted_probabilities_B, 'r', markerfmt='ro', basefmt=" ", linefmt='r-', label='Investment B')
ax[1, 0].set_title('Probability Density Function (PDF)')
ax[1, 0].set_xlabel('Profits ($M)')
ax[1, 0].set_ylabel('Probability')
ax[1, 0].set_ylim(-0.01, 0.5)
ax[1, 0].legend(loc=2)

# Plot the CDF for investment B
ax[1, 1].step(sorted_profits_B, cdf_B, 'r', where='post', label='Investment B')
ax[1, 1].set_title('Cumulative Distribution Function (CDF)')
ax[1, 1].set_xlabel('Profits ($M)')
ax[1, 1].set_ylabel('Cumulative Probability')
ax[1, 1].set_ylim(-0.01, 1.1)
ax[1, 1].legend(loc=2)

# Tight layout for neatness
plt.tight_layout()

# Save the figure as a PDF
plt.savefig('Problem4_Nguyen Xuan Binh_887799.pdf')

# Show the plot
plt.show()

```
