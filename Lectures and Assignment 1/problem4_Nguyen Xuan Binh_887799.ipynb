```python
import numpy as np
import matplotlib.pyplot as plt

# Define the scenarios and their probabilities
scenarios = np.array([1, 2, 3, 4, 5, 6])
probabilities = np.array([0.05, 0.15, 0.30, 0.35, 0.10, 0.05])

# Define the profits for each investment and scenario
profits_A = np.array([-10, 0, 10, 20, 20, 50])
profits_B = np.array([-5, -5, 0, 30, 40, 70])

# Sort the profits for the cumulative distribution function (CDF)
sorted_indices_A = np.argsort(profits_A)
sorted_indices_B = np.argsort(profits_B)

sorted_profits_A = profits_A[sorted_indices_A]
sorted_probabilities_A = probabilities[sorted_indices_A]

sorted_profits_B = profits_B[sorted_indices_B]
sorted_probabilities_B = probabilities[sorted_indices_B]

# Calculate the CDF for both investments
cdf_A = np.cumsum(sorted_probabilities_A)
cdf_B = np.cumsum(sorted_probabilities_B)

# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot the PDF for both investments
ax[0].bar(profits_A, probabilities, width=5, label='Investment A', alpha=0.7)
ax[0].bar(profits_B + 2.5, probabilities, width=5, label='Investment B', alpha=0.7)  # Offset for clarity
ax[0].set_title('Probability Density Function')
ax[0].set_xlabel('Profits ($M)')
ax[0].set_ylabel('Probability')
ax[0].legend()

# Plot the CDF for both investments
ax[1].step(sorted_profits_A, cdf_A, where='post', label='Investment A')
ax[1].step(sorted_profits_B, cdf_B, where='post', label='Investment B')
ax[1].set_title('Cumulative Distribution Function')
ax[1].set_xlabel('Profits ($M)')
ax[1].set_ylabel('Cumulative Probability')
ax[1].legend()

# Tight layout for neatness
plt.tight_layout()

# Save the figure as a PDF
plt.savefig('investment_profit_distributions.pdf')

plt.show()
```
