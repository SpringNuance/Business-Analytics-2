# Risk analysis in product development (R)

```r
# generate an empty vector on which the simulated profits will be appended to
P <- c()

for (i in 1:20000) { # generate 20 thousand simulations
    
    # c1 that should be distributed according to the given discrete distribution
    c1vals <- c(43,44,44,45,45,45,45,46,46,47) # we repeat e.g. 45 four times because it has probability .4
    c1 <- c1vals[runif(1,1,11)]

    # c2 that should be distributed uniformly
    c2 <- runif(1,80,100) 

    # D that should be distributed normally
    D <- rnorm(1, mean=15000, sd=4500)    

    # calculate profit for this simulation run
    newP <- (249 - c1 - c2)*D - 10^6

    # add the calculated profit to the vector of profits
    P <- c(P, newP)
}
```

```r
# histogram of profits
hist(P)
```
