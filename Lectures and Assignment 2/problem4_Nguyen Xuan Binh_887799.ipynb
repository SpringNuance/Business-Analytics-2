To estimate the utility function for the profit range [15 M€, 85 M€], we will use the certainty equivalents provided by the CEO and apply the concept of expected utility. The utility function will be scaled such that the utilities of the least preferred profit (15 M€) and the most preferred profit (85 M€) are 0 and 1, respectively.

The CEO's responses give us the following certainty equivalents:

A 50-50 gamble between 15 M€ and 85 M€ is equivalent to a certain profit of 32 M€.

A 50-50 gamble between 32 M€ and 85 M€ is equivalent to a certain profit of 49 M€.

A 50-50 gamble between 15 M€ and 32 M€ is equivalent to a certain profit of 21 M€.

A gamble with probability 0.9 for 85 M€ and 0.1 for 15 M€ is equivalent to a certain profit of 68 M€

From these data points, we can calculate the utility of each profit amount using the expected utility theory. The utility function U(x) can be approximated by solving these equations:


0.5 x U(15) + 0.5 x U(85) = U(32)

=> U(32) = 0.5 x 0 + 0.5 x 1 = 0.5

0.5 x U(32) + 0.5 x U(85) = U(49)

=> U(49) = 0.5 x 0.5 + 0.5 x 1 = 0.75

0.5 x U(15) + 0.5 x U(32) = U(21)

=> U(21) = 0.5 x 0 + 0.5 x 0.5 = 0.25

0.9 x U(85) + 0.1 x U(15) = U(68)

=> U(68) = 0.9 x 1 + 0.1 x 0 = 0.9

```python
from scipy.optimize import fsolve
import numpy as np
import matplotlib.pyplot as plt

# Define the equations based on the CEO's responses
def equations(vars):
    u32, u49, u21, u68 = vars
    eq1 = 0.5 * 0 + 0.5 * 1 - u32  # U(32) from 50-50 gamble between 15 and 85
    eq2 = 0.5 * u32 + 0.5 * 1 - u49  # U(49) from 50-50 gamble between 32 and 85
    eq3 = 0.5 * 0 + 0.5 * u32 - u21  # U(21) from 50-50 gamble between 15 and 32
    eq4 = 0.9 * 1 + 0.1 * 0 - u68  # U(68) from gamble with 90% for 85 and 10% for 15
    return [eq1, eq2, eq3, eq4]

# Initial guesses for the utilities
initial_guesses = [0.5, 0.5, 0.5, 0.5]

# Solve the equations
u32, u49, u21, u68 = fsolve(equations, initial_guesses)

# Profits corresponding to the calculated utilities
profits = [15, 21, 32, 49, 68, 85]
utilities = [0, u21, u32, u49, u68, 1]

# Plotting the utility function
plt.figure(figsize=(10, 6))
plt.plot(profits, utilities, marker='o')
plt.title('Utility Function for Profits')
plt.xlabel('Profit (M€)')
plt.ylabel('Utility')
plt.xticks(profits)
plt.yticks(utilities)
plt.grid(True)
plt.show()

u32, u49, u21, u68  # Display the calculated utilities
```
